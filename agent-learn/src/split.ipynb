# HTML Text Splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html_string = """
<!Doctype html>
<html>
    <head>
        <title>My HTML Document</title>
    </head>
    <body>
        <h1>Heading 1</h1>
        <p>This is a paragraph under heading 1.</p>
        <h2>Heading 2</h2>
        <p>This is a paragraph under heading 2.</p>
        <h3>Heading 3</h3>
        <p>This is a paragraph under heading 3.</p>
    </body>
</html>
"""
```

```python
headers_to_split_on = [
    ("h1", "Heading 1"),
    ("h2", "Heading 2"),
    ("h3", "Heading 3")
]
header_splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
html_header_splits = header_splitter.split_text(html_string)
html_header_splits[0]
```
