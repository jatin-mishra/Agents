# Line, Plane and Hyperplane

```python
# w1 x1 + w2 x2 + w0 = 0  -> line
# w1 x1 + w2 x2 + w3 x3 + w0 = 0 -> plane
# w1 x1 + w2 x2 + w3 x3 + w4 x4 + ....  + w0 = 0 -> hyperplane in higher dimensions

# Vector notation
# w0 + [w1, w2, w3, w4, ..., wn] . [x1, x2, x3, x4, ..., xn] = 0
# w0 + wT x = 0 -> decision boundary in higher dimensions
# w -> (n x 1)
# x -> (n x 1)

# if passing through origin -> wT x = 0 -> decision boundary through origin


# if wT x = 0 (passing through origin)
# w . x = |w| |x| cos(theta) = 0
# w . x = 0 -> theta = 90 degrees -> orthogonal (perpendicular) -> decision boundary is orthogonal to weight vector w
# any point x (vector) on plane satisfies w . x = 0 -> decision boundary is orthogonal to weight vector w
# w here is vector perpendicular to the plane x makes the equation wT x = 0 satisfied (if plane passes through origin)
```

```python
# Intercept form of hyper plane 
# x1/a + x2/b + x3/c + x4/d + ... = 1
# w1 x1 + w2 x2 + w3 x3 + w4 x4 + ....  + w0 = 0 -> w1 x1 + w2 x2 + w3 x3 + w4 x4 + ....  = -w0
# x1/( -w0/w1) + x2/(-w0/w2) + x3/(-w0/w3) + x4/(-w0/w4) + ... = 1
# a = -w0/w1, b = -w0/w2, c = -w0/w3, d = -w0/w4 -> intercepts on each axis

# given any dimentions vector w and x 
# wT x + b = 0 => intercept of ith dimension = -b/w_i
```
