# TF-IDF

- term frequency: TF(wi, rj) = ((# of times wi occurs in rj) / (total # of words in rj))
- inverse document frequency(wi, {r1, r2, r3, .... rn}): log((#of documents) / (# of documents which has wi))


## Formula
- Tf(w2, r1) * IDF(w2, {ri's})


## Why do we use log(N/ni) for IDF?
- Zipf's law: c type graph (power law distribution) of words and frequencies in english
