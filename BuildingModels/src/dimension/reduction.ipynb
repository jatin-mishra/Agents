# PCA (Principle Component Analysis)

find a frame in which direction, spread is the max (meaning preserving most of the information)

## Problem:
find u1 such that var{ proj{on u1}{of x_i}} is maximal.

var(transpose(u1) * x_i)

<proof yourself>

<for standardized columns>

problem is equivalent to solving:

find u1 which maximizes  ((1/n) * sum(transpose(u1) * xi) ** 2)

or minimize distance of all points from unit vector

u1 is eigen vector for which eigen value is largest. also using eigen value, it tells how much %age of data is retained.




## Summary
- find eigen vectors
- pick the ones preserving the most
- take projection of data vectors on those vectors depending on number of dimensions you want.

## Limitations of PCA
- all equal eigen values -> high loss
- cluster might get killed (if groups can be separated based in removed dimensions)
- if data follows like sin graph (periodic) in one direction but when we remove one dimension then that pattern is lost




```python
from sklearn.preprocessing import StandardScaler
data = [[1,2], [3,4]]
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data)
```

```python
import numpy as np
covar_matrix = np.matmul(standardized_data.T, standardized_data)
print(covar_matrix.shape)
```

```python
from scipy.linalg import eigh

values, vectors = eigh(covar_matrix, eigvals_only=(0))

vectors = vectors.T

print(vectors)
print(vectors.shape)
```
