# Measuring performance metrics
- accuracy (simple percentage)
- confusion matrix 
- precision
- recall
- F1-score
- receiver operating characteristic curver (ROC curve) & Area under curve (AUC)
- Log loss
- R^2 or Coefficient of determination (for regression model)

## Confusion matrix (Predicated vs Actual)
- 2D matrix: one side has actual values and other side has predicted values
- principle diagnals must be high, off diagnals must be small

- TPR = TP / P => divided by number actual positives
- TNR = TN / N => divided by number actual negatives
- FPR = FP / N => divided by number actual negatives
- FNR = FN / P => divided by number actual positives

## Precision
= TP / (TP + FP) 
=> here we divided by number of positives predicted as per model

## Recall
= TP / P
=> divided by number actual positives

## F1 Score
= 2 * (precision * recall) / (precision + recall)
=> inverse of simple avg of inverse of recall and precision

## Receiver operating characteristic curve & AUC (Only useful for binary classification but can be extended)
- for different threshold prob, build graph of (TPR vs FPR)


### AUC
- can be high for dumb model
- not dependent on y scores but only on ordering
- AUC between 0.5 to 1 the okay but smaller than 0.5 is worse than even random prediction (just swap the class label)


## Log Loss
= (-1/n)(sum from i = 1 to n)(  (log(pi) * yi)    +   ((1 - yi) * log(1 - pi))  )

### For multi class
= -1/n * sum(i from 1 to n) sum(j from 1 to n) (yij * log(pij))
