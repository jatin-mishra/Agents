# Naive Bayes
- Classification
- Probability based


## Conditional Probability
- P(A | B) : prob of A given B [ A, B are random vairables]
- P(A & B) / P(B)

## Independent Events & Mutual Exclusive events
- P(A|B) = P(A), P(B|A) = P(B) => independent events
- P(A|B) = P(B|A) = 0 => mutually exclusive events

# Bayes Theorem
P(A|B) = ( P(B|A) * P(A) ) / P(B) if P(B) != 0

# Naive Bayes Algorithm
- to find P(C_x | x) where x = {x1, x2, ...., xn}
- P(C_k | x) = ( P(C_k) * P(x | C_k) ) / P(x)
- denominator is constant so ignore
- Neumerator = P(C_k intersection x) = P(C_k, x) 
- for whichever class neumerator comes max <- will be the answer.
- x is n dimensional vector x = <x1, x2, x3, ...., xn> => P(C_k, x) = P(C_k, x1, x2, x3, x4) = P(x1, x2, x3, ..., C_k)
- = P(x1 | {x2, x3, ... C_k}) * P(x2, x3, ... C_k)
- = P(x1 | {x2, x3, ... C_k}) * 
        P(x2 | {x3, ... C_k}) * 
        P(x3 | {x4, ... C_k}) * 
        ... * 
        P(xn | C_k) * 
        P(C_k)
- **Assume all features are independent**
- = proportional to P(C_k) * multiplication_of(P(xi | C_k))
- = (1 / (Z=P(x))) * P(C_k) * multiplication_of(P(xi | C_k))


## Training
- Time Complexity: O( n d c)

# Laplace smoothing
- Scenario : what if some word comes which was not in training data and hence prob might become zero
- P(f1 | y = 1) = (0 + alpha) / (n1 + alpha * k)
- k = distinct values that f1 can take

- as alpha increases => moving likelihood prob to uniform distribution
- better to use log of probabilities

# Bias - Variance trade off

- alpha determines underfitting and overfitting.

# Feature importance and interpretability
- Feature importance can be obtained from model, words with max P(wi | y = 1) or P(wi | y = 0)
- Interpretability: has prob of class x is natural to get.
 
# Imbalance data
- Upsampling or Downsampling
- Drop P(y = 1) pr P(y = 0) part of prob calculation
- alpha impacts more minority class and less on majority class

# Outlier
- if word(x) comes lesser than a threshold then drop it.
- or use laplace smoothing

# Missing data
- text data: there is no missing issue
- categorical feature (consider NaN as category)
- numerical features: standard imputation (Gaussian naive bayes)

# Handling numerical features

- and use formula to calculate the value (as per distribution)
