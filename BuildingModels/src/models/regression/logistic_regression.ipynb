# Logistic Regression

- **Assumption**: classes are almost/perfectly linearly separable
- If my data is linearly separable  
- hyper-plane: transpose(w) x + b 

## Classifier
- if transpose(w) x > 0 => pos
- else if transpose(w) x < 0 => neg 
- assumption: w is unit vector perpendicular to plane
- yi transpose(w) x > 0 means working fine. (assuming pos means +1 and neg means -1 label)
- solve: maximize{w} sum{i from 1 to n} [yi transpose(w) x]

 
## Outlier
- if one outlier is very far from most points then if it has [ (yi transpose(w) outlier) < 0] and is big value then good plane might get rejected just because of that outlier.

### Handle outlier
- **Squashing**: is signed distance is small then keep it otherwise convert it to some smaller value
- **sigmoid function**: this functional tapers off big values -> f(x) = 1 / (1 + e ^ (-x))
- new problem to solve: argmax{w} sum{i from 1 to n} [ sigmoid-fxn( yi transpose(w) xi) ]


### Probabilistic interpretation
- it's like prob (0 to 1)
- x at -inf => prob = 0, x at inf => prob = 1
